```python
def calc_sales_tax(price, tax_rate=0.13):
  tax_amt = price * tax_rate
  return tax_amt
```

```python
def calc_total_bill(price, tax_rate=0.13, tip_rate=0.2):
  tax = price * tax_rate
  tip = price * tip_rate
  total_price = price + tax + tip
  return total_price
```

```python
calc_total_bill(100)
```

```python
calc_total_bill(tip_rate=0.22, price=100)
```
