```python
filter_list = ['999', '36 FINCH WEST', '35 JANE', '52', '41 KEELE']

delays_w_reasons = delays_w_reasons.loc[~delays_w_reasons['line'].isin(filter_list)]
delays_w_reasons['line'].unique()
```

```python
delays_w_reasons['line'] = (delays_w_reasons['line']
                            .str.replace('YUS', 'YU'))
```

```python
delays_w_reasons['line'].unique()
```

```python
delays_w_reasons['line'].value_counts()
```

```python
yubd_list = ['YONGE/UNIVERSITY/BLOOR',
             'YU / BD',
             'YU & BD',
             'YU & BD LINES']

delays_w_reasons.loc[delays_w_reasons['line'].isin(yubd_list), 'line'] = 'YU/BD'
```

```python
thefts_joined = pd.read_csv('../data/bike_thefts_joined.csv')
neighbourhoods = pd.read_csv('../data/neighbourhoods.csv')
```

```python
thefts_joined.info()
```

```python
thefts_joined['occurrence_date'] = pd.to_datetime(thefts_joined['occurrence_date'])
thefts_joined['report_date'] = pd.to_datetime(thefts_joined['report_date'])
```

```python
neighbourhoods.head()
```

```python
neighbourhoods = neighbourhoods.loc[neighbourhoods['neighbourhood'] != 'City of Toronto']
```

```python
# finding the % of commuters for each mode
def calc_pct(mode):
    return round(mode/neighbourhoods['total_commuters'], 3)

pct_cols = ['pct_drive', 'pct_cp', 'pct_transit', 'pct_walk']
neighbourhoods[pct_cols] = neighbourhoods.loc[:, 'drive':'walk'].apply(calc_pct)
```

```python
neighbourhoods.head()
```

```python
#jupyter-specific 
%matplotlib inline 

import matplotlib.pyplot as plt
```

```python
plt.plot(neighbourhoods['pop_2016'], neighbourhoods['bike'])
```

```python
plt.scatter(neighbourhoods['pop_2016'], neighbourhoods['bike'],
            marker='s', #square marker
            facecolor='#fb1', # set the color the marker's face
            edgecolor='k') #marker's edge or border is set to black
```

```python
plt.figure()
plt.scatter(neighbourhoods['pop_2016'],
            neighbourhoods['drive'],
            label='Driving')
plt.figure()
plt.scatter(neighbourhoods['pop_2016'],
            neighbourhoods['bike'],
            label='Cycling')
plt.legend()
```

```python
plt.grid()
```

```python
# object-oriented approach to plotting

fig, ax = plt.subplots()

type(fig)
```

```python
type(ax)
```

```python
drivers  = ax.scatter(neighbourhoods['pop_2016'], neighbourhoods['drive'])
cyclists  = ax.scatter(neighbourhoods['pop_2016'], neighbourhoods['bike'])
total  = ax.scatter(neighbourhoods['pop_2016'], neighbourhoods['total_commuters'])
fig
```

```python
ax.set_title('Neighbourhood Population vs Commuter Population')
ax.set_xlabel('Population, 2016')
ax.set_ylabel('Commuters')
ax.set_axisbelow(True)
ax.grid(alpha=0.7)
fig
```

```python
ax.legend([drivers, cyclists, total], ['Drivers', 'Cyclists','Total Commuters'],
          bbox_to_anchor=(0, 1),
          loc='upper left')

fig
```

```python
# modifying axis ticks

import matplotlib.ticker as tick
```

```python
ax.xaxis.set_major_formatter(tick.StrMethodFormatter('{x:,.0f}'))
ax.yaxis.set_major_formatter(tick.StrMethodFormatter('{x:,.0f}'))

# ax.xaxis --> accesses the x-axis of the plot represented by the ax object

# set_major_formatter() --> method that is used to set the formatting 
# for the major tick labels on the x-axis

# {x} --> placeholder for the value you want to format
# : --> acts as a delimiter, indicating that the formatting options for the value
# follow it
# , --> serves as a thousands separator
# .0 --> zero decimal places (rounds to the nearest integer)
# f --> format as a fixed-point number

fig
```
